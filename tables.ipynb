### Reading via read_csv
Pandas read_csv can also directly read from the link of a download see below

```python
# website link https://www.football-data.co.uk/data.php
data = pd.read_csv('https://www.football-data.co.uk/mmz4281/2324/E0.csv')

'https://www.football-data.co.uk/mmz4281/2324/E1.csv'
'https://www.football-data.co.uk/mmz4281/2324/EC.csv'
```

```python
dict = {}
for i in range(0,4):
  data = str('x'+str(i))
  dict[data] = i
print(dict)
```

```python
# Lets try to create a loop to automatically get the csv files
dict = {}
for num in range(0,4):
  data = pd.read_csv('https://www.football-data.co.uk/mmz4281/2324/E{}.csv'.format(num))
  dict[num] = data
```
